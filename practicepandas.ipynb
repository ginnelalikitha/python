```python
import pandas as pd
dataset={
    'cast':['matt','bonnie','elena','jermey','damon','stefan','tyler','klaus','elijah','rebekah','enzo','marcel'],
    'series':['vampire diaries','the originals','twilight','meg','got','arafta','my demon','first frost','sunlight','shore','zombie','israel'],
    'networth':[2,2.5,1,6,9,10,2.8,11,15,13.5,12,25],
    'properties':[5,7,2,9,10,18,45,32,68,10,14,23]

}
df=pd.DataFrame(dataset)
df
```

```python
import pandas  as pd
Mydataset={
    'Name':["Likitha","Elena","Zoya","Rithu","Ponnu","Firoj"],
    'Rollno':[12,15,43,62,29,20],
    'Marks':[45,50,49,40,35,38],
    'Subject':["Telugu","English","Maths","Science","Telugu","Social"],
    'properties':[5,7,2,9,10,18]
}
myvar = pd.DataFrame(Mydataset)
myvar
```

```python
merge=pd.merge(df,myvar,on='properties')
merge
```

```python
concat=pd.concat([df,myvar])
concat
```
