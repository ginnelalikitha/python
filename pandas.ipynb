#pivot


```
import pandas  as pd
store={
'product':['facewash','sunscreen','serum','kumkumadi thailam','facemask','moisturizer'],
'Brands':['hyphen','foxtale','dermaco','naturals','korean','nivea'],
        'price':[250,330,450,3000,120,400],
}
df = pd.DataFrame(store)
print(df)
```

```
df.groupby('product')['price'].sum()
```

```
df.groupby('Brands')['price'].sum()
```

```
#create pivot table
pivot=df.pivot_table(
    index='product',
    columns='Brands',
    values='price',
    aggfunc='sum',
    margins=True,
    margins_name='Total'
)
```

```
pivot
```

#crosstab

```
ct=pd.crosstab(df['product'],df['price'])
ct
```

#merge


```
import pandas as pd
store={
'product':['facewash','sunscreen','serum','kumkumadi thailam','facemask','moisturizer'],
'Brands':['hyphen','foxtale','dermaco','naturals','korean','nivea'],
        'price':[250,330,450,3000,120,400],
        'Marks':[45,50,49,40,35,38]
}
df = pd.DataFrame(store)
df
```

```
Mydataset={
    'Name':["Likitha","Elena","Zoya","Rithu","Ponnu","Firoj"],
    'Rollno':[12,15,43,62,29,20],
    'Marks':[45,50,49,40,35,38],
    'Subject':["Telugu","English","Maths","Science","Telugu","Social"]
}
myvar = pd.DataFrame(Mydataset)
myvar
```

```
merge=pd.merge(df,myvar,on='Marks')
merge
```

```
concatenate=pd.concat([df,myvar],ignore_index=True)
concatenate
```
